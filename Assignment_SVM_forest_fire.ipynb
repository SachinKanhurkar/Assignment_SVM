```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import preprocessing
```

```python
data= pd.read_csv("forestfires.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
label_encoder = preprocessing.LabelEncoder()
data['month']= label_encoder.fit_transform(data['month']) 
```

```python
data['day']= label_encoder.fit_transform(data['day']) 
```

```python
data['size_category']= label_encoder.fit_transform(data['size_category']) 
```

```python
data.head()
```

```python
data.head()
```

```python
n =min(data.shape)
        
```

```python
array = data.values
X = array[:,0:n]
Y = array[:,-1]
```

```python
n
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

### Grid Search CV

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 15, gamma = 0.5)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

### We classify the data for predict the size_Categorie the burned area of the forest ( Small = 0 , Large = 1).  We find the SVM based model with the 74.36% accuracy here with C = 15, gamma = 0.5 in RBF kernal fuction. Here, C is used for soft margin ( low c - allow soft margin, high c - punish model badly if make mistake), higher gamma signigy the more complexity in model by cosidering the more influcence of the dimensions on the deciosin bounday.
